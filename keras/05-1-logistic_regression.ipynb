```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import numpy as np
```

```python
import keras
```

```python
x_data = np.array([[1, 2],
                   [2, 3],
                   [3, 1],
                   [4, 3],
                   [5, 3],
                   [6, 2]])
y_data = np.array([[0],
                   [0],
                   [0],
                   [1],
                   [1],
                   [1]])
```

```python
model = Sequential()
model.add(Dense(1, input_dim = 2, activation = 'sigmoid'))
```

```python
sgd = SGD(lr = 0.1)
model.compile(loss = 'binary_crossentropy', optimizer = sgd, metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_data, y_data, epochs = 2000)
```

```python
print('2, 1:', 1 if model.predict(np.array([[2, 1]])) > 0.5 else 0)
print('6, 5:', 1 if model.predict(np.array([[6, 5]])) > 0.5 else 0)
```
